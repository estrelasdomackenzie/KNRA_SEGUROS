```python
print(combined_df)
```

```python
colunas_importantes = ['Agente  Causador  Acidente', 'Data Acidente' , 'Data Nascimento' , 'Sexo' , 'UF  Munic.  Acidente']
col_selecionadas = combined_df [colunas_importantes]
col_selecionadas.head()
```

```python
#str.strip()- retira os espaços em branco dentro da coluna
col_selecionadas['UF  Munic.  Acidente'] = col_selecionadas['UF  Munic.  Acidente'].str.strip()

#selecionamos o DF e a coluna que iremos tratar  e o que queremos extrair (!=  que dizer que só iremos extrair o que não for igual a "isso")
col_selecionadas = col_selecionadas[col_selecionadas['UF  Munic.  Acidente'] != '{ñ class}']

# iremos realizar a exclusão das linhas e o inplace true, significa que estamos mudando o DF original
col_selecionadas.reset_index(drop=True, inplace=True)

print(col_selecionadas)
```

```python
# o 1º formato de data estava como timestemp ( data e hora ) neste trecho estamos lidando com os erros dessa coluna
col_selecionadas['Data Nascimento'] = pd.to_datetime(col_selecionadas['Data Nascimento'], errors='coerce')

#transformando em forato de data
col_selecionadas['Data Nascimento'] = col_selecionadas['Data Nascimento'].dt.date

# novamente estamos "resetando o DF" e deixando essa versão tratada // quando o inplace está e FALSE ele não muda a base original e cria uma cópia para continuar as analises.
col_selecionadas.reset_index(drop=True, inplace=True)


print(col_selecionadas)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#ocorrencia por uf
ocorrencias_por_uf = col_selecionadas['UF  Munic.  Acidente'].value_counts()

# Plotar um gráfico de barras para visualizar as ocorrências por UF
plt.figure(figsize=(12, 6))
sns.barplot(x=ocorrencias_por_uf.index, y=ocorrencias_por_uf.values, palette='viridis')
plt.title('Número de Ocorrências por UF')
plt.xlabel('UF')
plt.ylabel('Número de Ocorrências')
plt.xticks(rotation=90)
plt.show()
```

```python
# ocorrências por Agente Causador e UF
ocorrencias_agente_uf = col_selecionadas.groupby(['Agente  Causador  Acidente', 'UF  Munic.  Acidente']).size().unstack()

# Verificar se a contagem de categorias não é muito alta
if ocorrencias_agente_uf.shape[0] > 20:
    ocorrencias_agente_uf = ocorrencias_agente_uf.head(20)  #   20 primeiros

# Configurar o gráfico
plt.figure(figsize=(14, 8))

# Plotar o gráfico
ocorrencias_agente_uf.plot(kind='bar', stacked=True, colormap='Set3', figsize=(14, 8))

# título e rótulos
plt.title('Ocorrências por Agente Causador e UF', fontsize=16)
plt.xlabel('Agente Causador', fontsize=12)
plt.ylabel('Número de Ocorrências', fontsize=12)

#
plt.xticks(rotation=45, ha='right')

# layout evitar superpposição
plt.tight_layout()

# mostrar o gráfico
plt.show()

```

```python
# ocorrências por sexo
ocorrencias_sexo = col_selecionadas['Sexo'].value_counts()

# graf. pizza
plt.figure(figsize=(8, 8))
plt.pie(ocorrencias_sexo.values, labels=ocorrencias_sexo.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])

# formato circulo
plt.title('Distribuição de Ocorrências por Sexo', fontsize=16)
plt.axis('equal')

# Mostrar o gráfico
plt.show()
```
